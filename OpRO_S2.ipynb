```python
#Problème 1 : min de Cmean

d = [6,4,6,3,2,7,1]
n = len(d) #nombre de tâches


#Modèle et variable
model = gp.Model()
x     = model.addMVar((n,n), vtype=gp.GRB.BINARY)

#C[i] = date de fin du job qui se trouve en position i !
C     = model.addVars(n, vtype=gp.GRB.CONTINUOUS) #dates de fin

#Contraintes
for i in range(n):
  model.addConstr(sum(x[i,:])==1)
for j in range(n):
  model.addConstr(sum(x[:,j])==1)


#Ecrivons la contrainte pour exprimer C[0]
#càd trouver la date de din du job se trouvant en position 0
somme = 0
for i in range(n):
  somme += d[i]*x[i,0]
model.addConstr(C[0]==somme)

#Pour les autres, C[1], C[2], etc
for j in range(1,n):
  somme = 0
  for i in range(n):
    somme += d[i]*x[i,j]
  model.addConstr(C[j]==C[j-1]+somme)

#Objectif
model.setObjective(C.sum(),gp.GRB.MINIMIZE)

#Lancement du solveur
model.optimize()

#Affichage des solutions
for j in range(n):
  indice = -1
  for i in range(n):
    if x[i,j].x > 0:
      indice = i
      break
  print(f"En position {j}, la tache {indice+1} associée est terminée en C = {C[j].x}")


```

```python
import gurobipy as gp

# Problème 1 : min de Lmax (minimiser le temps de fin maximal)
d = [6, 4, 6, 3, 2, 7, 1]  # durées des tâches
e = [15,6,9,23,20,30]
n = len(d)  # nombre de tâches

# Modèle et variables
model = gp.Model()
x = model.addMVar((n, n), vtype=gp.GRB.BINARY)

# C[j] = date de fin du job qui se trouve en position j
C = model.addVars(n, vtype=gp.GRB.CONTINUOUS)
lmax = model.addVar(vtype=gp.GRB.CONTINUOUS)

# Lmax = date de fin la plus tardive (max des C[j])
Lmax = model.addVar(vtype=gp.GRB.CONTINUOUS)

# Contraintes : chaque tâche est placée une seule fois
for i in range(n):
    model.addConstr(sum(x[i, :]) == 1)
for j in range(n):
    model.addConstr(sum(x[:, j]) == 1)

# Calcul de C[0]
somme = 0
for i in range(n):
    somme += d[i] * x[i, 0]
model.addConstr(C[0] == somme)

# Calcul des autres C[j]
for j in range(1, n):
    somme = 0
    for i in range(n):
        somme += e[i] * x[i, j]
    model.addConstr(C[j] == C[j - 1] + somme)

# Contraintes pour Lmax
for j in range(n):
  somme = 0
  for i in range(n):
    somme += e[i] * x[i, j]
  model.addConstr(lmax >= C[j]-somme)

# Objectif : minimiser Lmax
model.setObjective(lmax, gp.GRB.MINIMIZE)

# Résolution
model.optimize()

# Affichage des solutions
for j in range(n):
    indice = -1
    for i in range(n):
        if x[i, j].x > 0.5:
            indice = i
            break
    print(f"En position {j}, la tâche {indice+1} (durée {d[indice]}) se termine en C = {C[j].x}")

print(f"Valeur optimale de Lmax = {lmax.x}")


#MAUVAIS
```
